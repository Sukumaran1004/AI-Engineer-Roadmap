Manual Implementation

```python
def cosine_similarity(emb_a,emb_b):
    dot = np.dot(emb_a,emb_b)
    norm_a = np.linalg.norm(emb_a)
    norm_b = np.linalg.norm(emb_b)
    return dot / (norm_a * norm_b)
```

```python

import pandas
texts = [
    "car",
    "vehicle",
    "automobile",
    "truck",
    "banana",
    "apple",
    "fruit"
]

sentence = model.encode(texts)
matrix = util.cos_sim(sentence, sentence)
data = pandas.DataFrame(matrix,columns=texts,index=texts)
print(data)
```

```python
query = "What are the advancements in NLP?"
documents = [
    "Machine learning enables advancements in NLP.",
    "Climate change is a pressing issue globally.",
    "Natural language processing allows machines to understand text."
]
```

```python
from sentence_transformers import SentenceTransformer,util
#Model 1 --> BGE
model_bge = SentenceTransformer("BAAI/bge-large-en-v1.5")
query_emb = model_bge.encode(query)
doc_emb = model_bge.encode(documents)

similarity = util.cos_sim(query_emb, doc_emb)
print("sequence length:",model_bge.get_sentence_embedding_dimension())
print(similarity)
```

```python
from sentence_transformers import SentenceTransformer

# Load the 'e5-large-v2' model
model_e5 = SentenceTransformer('intfloat/e5-large-v2')
query_emb = model_e5.encode(query)
doc_emb = model_e5.encode(documents)

print("sequence length:",model_e5.get_sentence_embedding_dimension())
similarity = util.cos_sim(query_emb, doc_emb)
print(similarity)
```

```python
from sentence_transformers import SentenceTransformer,util

model_mpnet = SentenceTransformer("sentence-transformers/multi-qa-mpnet-base-dot-v1")
query_emb = model_mpnet.encode(query)
doc_emb = model_mpnet.encode(documents)

print("sequence length:",model_mpnet.get_sentence_embedding_dimension())
similarity = util.cos_sim(query_emb, doc_emb)
print(similarity)
result = util.semantic_search(query_emb, doc_emb,top_k=1)
print(result)
print(documents[result[0][0]['corpus_id']])
```

```python
from sentence_transformers import SentenceTransformer,util
import numpy

model_qwen = SentenceTransformer("Qwen/Qwen3-VL-Embedding-2B")
query_embedding = model_qwen.encode(
        query,
        convert_to_tensor=True
    )

doc_embeddings = model_qwen.encode(
    documents,
    convert_to_tensor=True
)

similarities = util.cos_sim(
    query_embedding,
    doc_embeddings
)[0]

# Convert to float32 before converting to numpy to handle BFloat16
scores = similarities.cpu().float().numpy()

ranked = sorted(
    zip(documents, scores),
    key=lambda x: x[1],
    reverse=True
)

print("\nSimilarity Scores:\n")

for doc, score in ranked:
    print(f"{score:.4f} -> {doc}")

print("\nTop Retrieved Document:")
print(ranked[0][0])
```

TF-IDF Manual Implementation

```python
import numpy as np

corpus = [
    "purple is the best city in the forest",
    "there is an art to getting your way and throwing bananas on to the street is not it",
    "it is not often you find soggy bananas on the street"
]

def tf_idf(query):
    tf_scores = []
    document_count = 0

    for doc in corpus:
        words = doc.split()
        term_count = words.count(query)

        if term_count > 0:
            document_count += 1

        tf_scores.append(term_count / len(words))

    if document_count == 0:
        print(f'"{query}" not found in the corpus.')
        return

    idf = np.log10(len(corpus) / document_count)

    print(f"TF-IDF scores for '{query}':")
    for score in tf_scores:
        print(round(score * idf, 2))

tf_idf("forest")
```
